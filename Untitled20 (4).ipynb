```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

df = pd.read_csv("diabetes.csv")

print(df.head())

print(df.info())

print(df.describe())


mean_values = df.mean(numeric_only=True)
print("\nMean:")
print(mean_values)


median_values = df.median(numeric_only=True)
print("\nMedian:")
print(median_values)


mode_values = df.mode(numeric_only=True)
print("\nMode:")
print(mode_values)


# ===== Min & Max =====
min_values = df.min(numeric_only=True)
max_values = df.max(numeric_only=True)

print("\nMinimum values:")
print(min_values)

print("\nMaximum values:")
print(max_values)


percentiles = df.quantile([0.25, 0.50, 0.75])
print("\nPercentiles:")
print(percentiles)


Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)

IQR = Q3 - Q1

print("\nIQR:")
print(IQR)


plt.figure()
sns.boxplot(data=df)
plt.xticks(rotation=45)
plt.title("Boxplot for all features")
plt.show()


df.hist(figsize=(10,8))
plt.show()


plt.figure()
sns.heatmap(df.corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()

# ===== حفظ النتائج في ملف نصي =====

with open("results.txt", "w") as f:

    f.write("First 5 rows:\n")
    f.write(str(df.head()))

    f.write("\n\nInfo:\n")
    df.info(buf=f)

    f.write("\n\nDescription:\n")
    f.write(str(df.describe()))

    f.write("\n\nMean:\n")
    f.write(str(mean_values))

    f.write("\n\nMedian:\n")
    f.write(str(median_values))

    f.write("\n\nMode:\n")
    f.write(str(mode_values))

    f.write("\n\nMin:\n")
    f.write(str(df.min(numeric_only=True)))

    f.write("\n\nMax:\n")
    f.write(str(df.max(numeric_only=True)))

    f.write("\n\nPercentiles:\n")
    f.write(str(percentiles))

    f.write("\n\nIQR:\n")
    f.write(str(IQR))





import shutil

shutil.make_archive(
    "diabetes_analysis",
    'zip',
    '.'
)

```
